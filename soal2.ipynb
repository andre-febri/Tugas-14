```python
def cari_maksimum(arr, kiri, kanan):
    if kiri == kanan:
        return arr[kiri]

    tengah = (kiri + kanan) // 2

    maks_kiri = cari_maksimum(arr, kiri, tengah)
    maks_kanan = cari_maksimum(arr, tengah + 1, kanan)

    if maks_kiri > maks_kanan:
        return maks_kiri
    else:
        return maks_kanan


n = int(input("Jumlah data: "))
arr = list(map(int, input("Masukkan angka: ").split()))

hasil = cari_maksimum(arr, 0, n - 1)
print("Nilai maksimum =", hasil)
```
