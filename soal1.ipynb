```python
money = int(input("Jumlah uang: "))

coins = [100000, 50000, 20000, 10000, 5000, 2000, 1000]

total = 0

print("\nHasil Penukaran")

for coin in coins:
    count = money // coin
    money %= coin
    total += count
    print(f"{coin:6} : {count} lembar")

print("\nTotal lembar =", total)
```
