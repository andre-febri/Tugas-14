```python

n = int(input("Masukkan jumlah anak tangga: "))

if n == 1:
    print("Jumlah cara = 1")
elif n == 2:
    print("Jumlah cara = 2")
else:
    dp = [0] * (n + 1)

    dp[1] = 1
    dp[2] = 2

    for i in range(3, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]

    print("Jumlah cara =", dp[n])
```
